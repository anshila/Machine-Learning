#### Outlier Removal

```python
#box plot to identify outliers
fig = plt.figure(figsize=(30,30))
counter = 0

for col in df1.select_dtypes(include="number").columns:
    fig.add_subplot(4,4,counter+1)
    sns.boxplot(data=df1[col])
    plt.xlabel(col)
    counter = counter + 1
```

```python
import math
def whisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
for i in df1.select_dtypes(include="number").columns:
    lw,uw=whisker(df1[i])
    df1[i]=np.where(df1[i]<lw,uw,df1[i])
    df1[i]=np.where(df1[i]>uw,uw,df1[i])
```

```python
#box plot after removal of outliers
fig = plt.figure(figsize=(30,30))
counter = 0

for col in df1.select_dtypes(include="number").columns:
    fig.add_subplot(4,4,counter+1)
    sns.boxplot(data=df1[col])
    plt.xlabel(col)
    counter = counter + 1
```

#### ENCODING

```python
# Label Encoding

from sklearn.preprocessing import LabelEncoder

cols_to_encode=list(df1.select_dtypes(include=["object"]))
cols_to_encode

#Label_encoder is the object of LabelEncoder
label_encoder=LabelEncoder()

#Initialize a new dataframe to store the encoded columns
df_label=df1.copy()

#Encode each columns in cols_to-encode
for col in cols_to_encode:
    df_label[col]=label_encoder.fit_transform(df1[col])

print(df_label)
```

#### FEATURE SCALING

```python
from sklearn.preprocessing import MinMaxScaler

# Select numerical columns for scaling
numerical_columns = ['carlength', 'carwidth', 'carheight', 'enginesize', 'boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','price']

# Create a scaler object
scaler=MinMaxScaler()

#Fit the scaler on the data
scaler.fit(df[numerical_columns])

#Transform the data
scaled_data=scaler.transform(df[numerical_columns])

#Convert the scaled data to dataframe
scaled_df=pd.DataFrame(scaled_data,columns=numerical_columns)

# Initialize a new DataFrame to store after applying MinMaxScaler
df_car=df_label.copy()

#Replace the scaled columns in original dataframe
df_car[numerical_columns]=scaled_df

# Print the DataFrame with scaled columns replaced by original values
print(df_car)
```

### <span style='color:Green'>Model Creation</span>

### Data Splitting

```python
#Splitting dataset to features and variables
x = df_car.drop('price', axis=1)
y = df_car['price']
x
```

```python
y
```

```python
#split the dataset to training and test sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

We can use these training and testing sets for regression algorithms

### 1. LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
LinearModel=LinearRegression()
```

```python
#Train the model with the provided data
LinearModel.fit(x_train,y_train)
```

```python
LinearPredictedTest=LinearModel.predict(x_test)
LinearPredictedTest
```

```python
#plot the actual target variable against the predicted
plt.scatter(y_test,LinearPredictedTest,color='red',label='Actual')
plt.plot(y_test,y_test,color='blue',label='Prediction')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Linear Regression Predictions vs Actual Price')
plt.legend()
```

### **EVALUATION**

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print("Linear Regression")
LR_MAE=mean_absolute_error(y_test,LinearPredictedTest)
LR_MSE=mean_squared_error(y_test,LinearPredictedTest)
LR_RMSE=np.sqrt(LR_MSE)
LR_r2=r2_score(y_test,LinearPredictedTest)
print("Mean Absolute Error(MAE) : ",LR_MAE)
print("Mean Squared Error(MSE) : ",LR_MSE)
print("RMSE : ",LR_RMSE)
print("R Score : ",LR_r2)
```

```python
# lr.score(x_test, y_test) calculates the R-squared score on the test data .
LinearModel.score(x_test,y_test)
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
scores=cross_val_score(LinearModel,x_train,y_train,cv=5)
scores
```

```python
#Mean of score
scores.mean()
```
