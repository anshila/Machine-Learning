#### OUTLIER REMOVAL

```python
def whisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
#removing outliers using capping
for i in df.select_dtypes(include="number").columns:
    lw,uw=whisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i]) 
```

```python
#box plot after outlier treatment
fig = plt.figure(figsize=(30,30))
counter = 0

for col in df.select_dtypes(include="number").columns:
    fig.add_subplot(4,4,counter+1)
    sns.boxplot(data=df[col])
    plt.xlabel(col)
    counter = counter + 1
```

# CLUSTERING

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

### 1. K Means Clustering

When there are many clusters, determining the optimal value of K using a scatter plot can be challenging. In such cases, we can use an elbow plot to identify the best value for K.

#### Elbow Plot

```python
clustering_data = df.iloc[:,[0,1,2,3]]
clustering_data.head()
```

```python
km = KMeans(n_clusters = 3)
km.fit(clustering_data)
```

```python
clusters = clustering_data.copy()
clusters['Cluster_Prediction'] = km.fit_predict(clustering_data)
clusters.head()
```

```python
km.cluster_centers_
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(i)
    km.fit(clustering_data)
    wcss.append(km.inertia_)
np.array(wcss)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
ax = plt.plot(range(1,11),wcss,linewidth = 2, color = 'green',marker = '8')
plt.axvline(x=3,ls='--')
plt.ylabel('WCSS')
plt.xlabel('No of Clusters (k)')
plt.title('The Elbow Method', fontsize = 20)
plt.show()
```

We can see that K=3 from the Elbow plot.

#### Silhouette Coefficient

We can use the silhouette score to evaluate the quality of clustering. The silhouette score measures how similar an object is to its own cluster (cohesion) compared to other clusters (separation). A higher silhouette score indicates better clustering.

```python
s=[]
k_rng = range(2,11)
for k in k_rng:
    km=KMeans(n_clusters=k)
    km_pred=km.fit_predict(clustering_data)
    s.append(silhouette_score(clustering_data,km_pred))
s
```

These are the silhouette scores for different values of K 

```python
silhouette_score(clustering_data,km_pred)
```

```python
plt.xlabel('K')
plt.ylabel('silhouette score')
plt.plot(k_rng,s,color='blue')
plt.title('Silhouette Score vs. Number of Clusters')
```

```python
optimalkmeans = KMeans(n_clusters = 2,init = 'k-means++', max_iter = 300,n_init = 10, random_state = 0)
y_kmeansbest = optimalkmeans.fit_predict(clustering_data)
```

```python
silhouette_score(clustering_data, y_kmeansbest)
```

### 2. Heirarchical Clustering


```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(18, 6))
# Calculate linkage using Ward's method and create the dendrogram
dendro = shc.dendrogram(shc.linkage(clustering_data, method="ward"))
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
```

- Identify clusters by looking at the vertical lines in the dendrogram. The height of the lines represents the distance or dissimilarity between clusters.
- To determine the optimal number of clusters, look for the longest vertical line that can be cut horizontally without intersecting other vertical lines (forming distinct clusters).

```python
plt.figure(figsize=(18, 6))
dendro = shc.dendrogram(shc.linkage(clustering_data, method="ward"))
plt.hlines(y=4,xmin=0,xmax=500,lw=3,linestyles='--')
plt.text(x=20,y=1.5,s='Horizontal line crossing 2 vertical lines')
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
```
